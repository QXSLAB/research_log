# 使用Parallelism in skorch


## 环境配置
pip install --user dask distributed --upgrade
%不加--user会报没有权限

pip install --user bokeh --upgrade



## 网络搭建

### 在第一个cmd窗口（启动scheduler，调度器）
dask-scheduler
%启动调度器

### 在第二个cmd窗口（启动第一个worker）
set CUDA_VISIBLE_DEVICES=0
%使这个worker只能看到第一个GPU
dask-worker 127.0.0.1:8786 --nthreads 1
%启动第一个worker，注册到调度器中

### 在第三个cmd窗口（启动第二个worker）
set CUDA_VISIBLE_DEVICES=1
%使这个worker只能看到第一个GPU
dask-worker 127.0.0.1:8786 --nthreads 1
%启动第二个worker，注册到调度器中
