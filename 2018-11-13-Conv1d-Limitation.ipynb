# conv1d的局限性

conv1d更加像是一种全连接的神经网络，因为现在的数据集使用的是8倍过采样，而conv1d使用的是1x3的卷积核，其响应域近似为波形上的一个点。
conv1d没有进行pooling，无法扩展其响应域，学习更高层特征。
conv1d没有捕捉很长的特征

捕捉长特征可提高噪声鲁棒性
各种调制格式的信号在星座点空间分布以及星座点跃迁规则两方面存在差异，这导致了不同调制格式信号的波形分布不同。噪声会破坏波形的局部，但是由于噪声往往是零均值、固定方差的，所以噪声并不会影响波形的大尺度分布。

conv1d观察到的是许多离散的小尺度特征的随机混合，通过统计这些小尺度特征的出现次数来分辨不同的调制信号。但是无法有效地对这些小尺度特征之间的时间关联性进行建模。小尺度特征及其之间的时序关联性构成了大尺度特征。而信号的特征往往体现在大尺度特征中。

从信噪比的角度讲，更长的信号波形可以带来更多的信号特征。由于信号特征是相关的，噪声是不相关的，所以更长的信号波形可以提高信噪比。
